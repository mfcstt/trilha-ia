## Tratamento de Dados

```python
# Transformar coluna TotalCharges de String para Float - Primeira abordagem com astype
df_contracts['TotalCharges'] = df_contracts['TotalCharges'].astype(float)
```

```python
# Segunda Abordagem utilizando to_numeric
df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'])
```

```python
# Terceira abordagem utilizando to_numeric com coerce
df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'],errors='coerce')
df_contracts.info()
```

Após a tranformação da coluna TotalCharges do DataFrame Contracts, a coluna passou a ter onze valores ausentes (missing value)

## Renomeação de Colunas

```python
df_customers.info()
```

```python
# Renomear coluna SeniorCitizen do DataFrame Customers - Primeira Abordagem
df_customers.rename(columns={'SeniorCitizen': 'Above65yo'},inplace=True)
```

```python
# Segunda abordagem utilizando lista com o objetivo de modificar todos os nomes da coluna
df_customers.columns = ['idCliente', 'Genero', 'Mais65Anos', 'Status', 'TemDependentes']
df_customers.info()
```

## Unificar Datasets

```python
# Visualizar quantidade de registros por DataFrame
len(df_contracts)
```

```python
len(df_services)
```

```python
len(df_customers)
```

```python
# Renomear colunas
df_services.rename(columns={'customerID': 'idCliente'}, inplace=True)
df_services.info()
```

```python
# Unificar Dataframe de Customers com Services, criando um terceiro DataFrame
df_temp = df_customers.merge(df_services,on=['idCliente'])
df_temp.info()
```

```python
df_temp.head(5)
```

```python
# Unificar df_temp com contracts, usando colunas de junção com nomes distintos
df_churn_temp = df_temp.merge(df_contracts, left_on=['idCliente'], right_on=['customerID'])
df_churn_temp.info()
```

```python
# Unificar dataframes com colunas de junção com nomes distintos
df_churn = df_customers.merge(df_services, on=['idCliente']).merge(df_contracts,left_on=['idCliente'],right_on=['customerID'])
df_churn.info()
```

```python
df_churn.drop(['customerID'],axis=1, inplace=True)
df_churn.info()
```
