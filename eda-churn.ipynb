## Análise Bivariada

```python
# Tabela de contingência
pd.crosstab(df_churn.Churn,df_churn.Contract, margins= True, margins_name='Total')
```

```python
# Proporção
pd.crosstab(df_churn.Churn,df_churn.Contract, normalize='index', margins= True, margins_name='Total')
```

Ao analisar o público total, os clientes com contrato mensal representam 55%. No entanto, ao focarmos apenas nos clientes que encerraram o serviço, observamos que 88% deles possuíam contrato mensal. Essa constatação sugere uma correlação significativa entre a opção de contrato mensal e o fenômeno de abandono.

### Correlação de variáveis qualitativas


Conduzir um teste de hipótese Chi-Square ou Qui-Quadrado de Pearson:

Durante o teste, formulamos duas hipóteses:

H0 (Hipótese Nula): As duas variáveis são independentes.

H1 (Hipótese Complementar): As duas variáveis são dependentes.

Ao observar que a probabilidade de H0 é menor que 0.05 (p-value), rejeitamos a hipótese nula e aceitamos a hipótese complementar para prosseguir com a análise.

```python
df_crosstab_chrun_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Cálculo de Scores e P-value

from scipy.stats import chi2_contingency
```

```python
# Tupla - Lista imútavel
chi_scores_churn_contract = chi2_contingency(df_crosstab_chrun_contract)
chi_scores_churn_contract
```

```python
# Series
score_churn_contract = pd.Series(chi_scores_churn_contract[0])
pvalue_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar resultado sem notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
#DataFrame com resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': score_churn_contract, 'p-value': pvalue_churn_contract})
df_chi_scores_churn_contract
```

Com base no teste estatístico (Qui-Quadrado), onde o valor-p (p-value) é menor ou igual a 0.05, rejeitamos a hipótese nula, indicando que as variáveis não são independentes. Além disso, um Qui-Quadrado elevado sugere uma correlação significativa entre as variáveis, reforçando a presença de uma associação robusta entre elas.

### Correlação entra variáveis qualitativas e quantitativas

### Método de Tukey 

O Método de Tukey, empregando o IQR (Intervalo Interquartil), é uma abordagem útil quando lidamos com distribuições não normais. Para implementar esse método, é necessário definir limites com base no IQR, que é calculado como a diferença entre o 3º Quartil e o 1º Quartil.

```python
q1_TotalCharges_monthly = df_churn_monthly.TotalCharges.quantile(0.25)
q3_TotalCharges_monthly = df_churn_monthly.TotalCharges.quantile(0.75)
iqr_TotalCharges_monthly = q3_TotalCharges_monthly - q1_TotalCharges_monthly
iqr_TotalCharges_monthly
```

```python
# Cálculo de limites
limInf_TotalCharges_monthly = q1_TotalCharges_monthly - (iqr_TotalCharges_monthly * 1.5) 
limSup_TotalCharges_monthly = q3_TotalCharges_monthly + (iqr_TotalCharges_monthly * 1.5)
limInf_TotalCharges_monthly, limSup_TotalCharges_monthly
```

```python
df_churn_monthly[(df_churn_monthly.TotalCharges < limInf_TotalCharges_monthly) | (df_churn_monthly.TotalCharges > limSup_TotalCharges_monthly)]
```

### Método Zscore
Utilizado em contextos de distribuição normal, o método indica quantos desvios padrões um ponto de dados específico está distante da média. Qualquer número que ultrapasse 3 desvios padrões da média é considerado um valor atípico (outlier).

```python
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_monthly.TotalCharges))
z
```

```python
df_churn_monthly[(z > 3.0)]
```

## Conclusões

##### Resumo dos resultados
Com base nas análises realizadas no conjunto de dados do segmento de Telecom, algumas conclusões relevantes emergiram. A variável "tempo de contrato" (tenure) exibiu uma distribuição não normal, concentrando-se principalmente em contratos de curta e longa duração. Observou-se uma forte associação entre a escolha do contrato mensal e o fenômeno de abandono (Churn), com 88% dos clientes que encerraram o serviço possuindo contratos mensais. O teste estatístico Qui-Quadrado reforçou essa associação, indicando uma correlação significativa e rejeitando a hipótese nula de independência entre as variáveis. Além disso, a análise apontou uma correlação robusta entre o tempo de contrato (tenure) e o TotalCharges, validada por métodos estatísticos de correlação, como Pearson e Spearman.

##### Interpretação dos Resultados:
Esses resultados sugerem que a escolha do tipo de contrato e a duração do contrato têm impacto significativo no Churn. Clientes com contratos mensais são mais propensos a abandonar o serviço, e a duração do contrato influencia as despesas totais. A constatação de que clientes com contratos mensais são mais propensos ao Churn pode ser explicada pela menor obrigação contratual. A liberdade de encerrar o serviço sem penalidades financeiras pode estar influenciando essa escolha.

##### Implicações dos Resultados:
As descobertas podem orientar estratégias de retenção de clientes, destacando a importância de incentivar contratos de longo prazo. A empresa pode considerar ofertas especiais para clientes com contratos mensais a fim de reduzir a taxa de Churn.

##### Limitações da Análise:
A análise se baseia em correlações e associações, não implicando causalidade. Outras variáveis não consideradas podem influenciar o Churn. Além disso, a falta de dados em TotalCharges para alguns clientes pode impactar a análise.

##### Sugestões para Futuras Análises:
Futuras análises podem explorar outras variáveis, como a satisfação do cliente, para uma compreensão mais abrangente do Churn. Investigar a eficácia de estratégias específicas de retenção também pode ser valioso.
